```python
import requests

# Collecting Exchange Rates to USD
url = 'https://v6.exchangerate-api.com/v6/b5c00762245c4b6467548e58/latest/USD'

# Making the request
response = requests.get(url)
currency_rates_usd = response.json()
```

```python
def convert_currency(origin_currency):
    if origin_currency in currency_rates_usd['conversion_rates']:
        return currency_rates_usd['conversion_rates'][origin_currency]
    else:
        pass
```

```python
def bin(x):
    if (x<=-480000):
        return 1
    elif (x >-480000) and (x <=-360000):
        return 2
    elif (x>-360000) and (x<=-240000):
        return 3
    elif (x>-240000) and (x<=-120000):
        return 4
    elif (x>-120000)  and (x<=0):
        return 5
    elif (x>0) and (x<= 120000):
        return 6
    elif (x>120000) and (x<= 240000):
        return 7
    elif (x>240000) and (x<= 360000):
        return 8
    elif (x>360000) and (x<= 480000):
        return 9
    elif (x>480000):
        return 10
```

```python
def bin(x):
    if (x >= -300000) and (x<= -240000):
        return 1
    elif (x >-240000) and (x <= -180000):
        return 2
    elif (x>-180000) and (x<=-120000):
        return 3
    elif (x>-120000) and (x<=-60000):
        return 4
    elif (x>-60000) and (x<=0):
        return 5
    elif (x>0) and (x<=60000):
        return 6
    elif (x>60000) and (x<=120000):
        return 7
    elif (x>120000) and (x<=180000):
        return 8
    elif (x>180000) and (x<=240000):
        return 9
    elif (x>240000) and (x<=300000):
        return 10
     
```
